```python
tourney_results = pd.read_csv('TourneyDetailedResults.csv')
```

Imports Seeding Data, cleans it (removes regional information)

```python
seeding = pd.read_csv('TourneySeeds.csv')
seeding['Seed'] = seeding['Seed'].map(lambda x: x.lstrip('WXYZ').rstrip('ab'))
seeding
```

```python
three_point = tourney_results[['Season', 'Wteam', 'Wfgm3','Lteam', 'Lfgm3']]
three_point = pd.merge(three_point, seeding, how='left', left_on=['Season', 'Wteam'], right_on = ['Season','Team'])
three_point.rename(columns={'Seed': 'Wseed'}, inplace=True)
del three_point['Team']
three_point = pd.merge(three_point, seeding, how='left', left_on=['Season', 'Lteam'], right_on = ['Season','Team'])
three_point.rename(columns={'Seed': 'Lseed'}, inplace=True)
del three_point['Team']
three_point
```
